```r
install.packages("babynames")
```

```r
library(babynames)
str(babynames)
```

```r
linnet_df <- babynames[babynames$name == "Linnet", ]
```

```r
plot(x = linnet_df$year,
    y = linnet_df$n,
    type = "l",
    main = "Popularity of the name Linnet")
```
