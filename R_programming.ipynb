```r
install.packages("babynames")
```

```r
library(babynames)
str(babynames)
```

```r
duane_df <- babynames[babynames$name == "Duane", ]
```

```r
plot(x = duane_df$year,
    y = duane_df$n,
    type = "l",
    main = "Popularity of the name Duane")
```
